```python

import csv
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# extracting csv data
csv_file = 'Top_Highest_Openings.csv'
df = pd.read_csv(csv_file)
```

```python
# creating line plot
fig = plt.figure()
ax = plt.axes()

df = df.sort_values(by = 'Date')
total_gross = df['Total Gross']
dates = pd.to_datetime(df['Date'], format='mixed')
# dates = df['Date']

plt.plot(dates, total_gross)
ax.set_title('Total Gross by Release Date')
ax.set_xlabel('Date')
ax.set_ylabel('Total Gross')

plt.show()
```

```python
# Scatter Plot
fig, ax = plt.subplots()
ax.scatter(dates, total_gross)

ax.set_title('Total Gross by Release Date')
ax.set_xlabel('Date')
ax.set_ylabel('Total Gross')

plt.show()
```

```python
# Bar plot
fig, ax = plt.subplots()

data = df['Total Gross']
ax.boxplot(data)
ax.set_title('Basic Box Plot')

plt.show()
```
